```python
import requests
from bs4 import BeautifulSoup
import urllib3

# warning 메시지 안나오게 하는 설정
urllib3.disable_warnings()

# 파라미터 종류가 많을 경우 딕셔너리 이용
param = {
    'isDetailSearch': 'N',
    'searchGubun': 'true',
    'viewYn': 'OP',
    'strQuery': '패션 인공지능',
    'order': '/DESC',
    'onHanja': 'false',
    'strSort': 'RANK',
    'iStartCount': 0,
    'sflag': 1,
    'isFDetailSearch': 'N',
    'pageNumber': 1,
    'resultKeyword': '패션 인공지능',
    'icate': 're_a_kor',
    'colName': 're_a_kor',
    'pageScale': '100',
    'isTab': 'Y',
    'query': '패션 인공지능'
}

response = requests.get("https://www.riss.kr/search/Search.do", params=param)
html = response.text
soup = BeautifulSoup(html, "html.parser")
articles = soup.select('.srchResultListW > ul > li')

# 헤더가 필요한 경우
header = {
    'User-Agent': 'Mozilla/5.0',
    'Referer': 'https://www.riss.kr/search/detail/DetailView.do?p_mat_type=1a0202e37d52c72d&control_no=dcd1a8e873c0ca787ecd42904f0c5d65&keyword='
}

for article in articles:
    title = article.select_one('.title > a').text
    link = 'https://www.riss.kr' + article.select_one('.title > a').attrs['href']

    # 상세 페이지 요청
    subResponse = requests.get(link, headers=header, verify=False)
    subHtml = subResponse.text
    subSoup = BeautifulSoup(subHtml, "html.parser")

    # 2. 텍스트를 기준으로 찾기
    # 발행기관의 바로 옆의 태그를 찾아라
    press = subSoup.find('span', string='발행기관').find_next_sibling().text
    year = subSoup.find('span', string='발행연도').find_next_sibling().text
    
    if subSoup.find('span', string='주제어'):
        keywords = subSoup.find('span', string='주제어').find_next_sibling().text.split(';')
        tags = [keyword.strip() for keyword in keywords]
    else:
        tags = []
    print(tags)
```
